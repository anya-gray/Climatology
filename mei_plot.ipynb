```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# plotting the MEI data for comparison
# reading in the csv file and extracting the data
with open('mei.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)

    # convert strings to numbers, collect everything in a list of lists
    headers = next(reader)
    data_list = [[float(item) for item in row] for row in reader if row]

# convert to numpy array for convenient indexing
data = np.array(data_list)
plt.figure(figsize=(12,3.5))
plt.plot(data[:])
plt.ylabel("Value")
plt.xlabel("Point in Time")
plt.title("MEI Data")

plt.show()
```
